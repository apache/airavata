## AlphaFold2

```python
exp = md.AlphaFold2.initialize(
    name="AlphaFold2_test",
    input_seq="data/alphafold2/Abeta42_Dimer.fasta",
    max_template_date="2023-09-18",
    model_preset="multimer",
    multimers_per_model=6,
)
exp.create_task(*ae.list_runtimes(cluster="login.expanse.sdsc.edu", category="gpu", group="Default", walltime=60))
ae.display(exp)
plan = exp.plan()
plan.launch()
plan.save_json("plan_alphafold2.json")
```
